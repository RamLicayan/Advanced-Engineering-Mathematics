> **1.4 Multiplication and division of complex numbers** 
>>1. Multiplication of complex numbers is 
achieved by assuming all quantities involved are real and then using $\mathrm{j}^2 = \mathrm{-1} $ to simplify.
2. The complex conjugate of a complex number is obtained by changing the sign of the imaginary part. Hence the complex conjugate of a + jb is a − jb. The product of a complex number and its complex conjugate is always a real number.
3. Division of complex numbers is achieved by multiplying both numerator and denominator by the complex conjugate of the denominator.

>**1.6 The polar form of a complex number**
>>1. Let a complex number z be x + jy as shown in
the Argand diagram of Fig. 20.4. Let distance
OZ be r and the angle OZ makes with the positive
real axis be θ.
>>\begin{equation}
\text { From trigonometry, } \mathrm{x}=\mathrm{r} \cos \theta \text { and }
\end{equation}
\begin{equation}
\begin{aligned}
& \mathrm{y}=\mathrm{r} \sin \theta \\
\text { Hence } \mathrm{Z}=\mathrm{x}+\mathrm{j} \mathrm{y} &=\mathrm{r} \cos \theta+\mathrm{j} \mathrm{r} \sin \theta \\
&=\mathrm{r}(\cos \theta+\mathrm{j} \sin \theta)
\end{aligned}
\end{equation}
>>$ \mathrm{Z}=\mathrm{r}(\cos \theta+\mathrm{j} \sin \theta)$ is usually abbreviated to
$\mathrm{Z}=\mathrm{r} \angle \theta$ which is known as the polar form of a complex number.
2. r is called the modulus (or magnitude) of Z and
is written as mod Z or |Z|.
r is determined using Pythagoras’ theorem on
triangle OAZ in Fig. 20.4,
\begin{equation}
{r=}\sqrt{x^2+y^2}
\end{equation}

> **Problem 8**

Determine, in polar form:

(a) $8 \angle 25^{\circ} \times 4 \angle 60^{\circ}$

(b) $3 \angle 16^{\circ} \times 5 \angle-44^{\circ} \times 2 \angle 80^{\circ}$


```
#code for Problem 8 (Method 1)

#import cmath, math and numpy
import cmath, math, numpy

#values of the angles in radians
a1 = math.radians(25)
a2 = math.radians (60)
b1 = math.radians (16)
b2 = math.radians (-44)
b3 = math.radians (80)

#converting from polar to rectangular form
za1 = cmath.rect(8,a1)
za2 = cmath.rect(4,a2)
zb1 = cmath.rect(3,b1)
zb2 = cmath.rect(5,b2)
zb3 = cmath.rect(2,b3)


#solving for a
a = za1*za2
m1 = abs(a)
d1 = numpy.degrees(cmath.phase(a))
print ("The answer for (a) is", m1, "∠", d1, "°" )

#solving for b
b = zb1*zb2*zb3
m2 = abs(b)
d2 = numpy.degrees(cmath.phase(b))
print ("The answer for (b) is", m2, "∠", d2, "°" )
```

>**1.8 Applications of Complex Numbers**
>>There are several applications of complex numbers
in science and engineering, in particular in electrical
alternating current theory and in mechanical vector
analysis.The effect of multiplying a phasor by j is to rotate
it in a positive direction (i.e. anticlockwise) on an
Argand diagram through 90◦ without altering its length.
Similarly, multiplying a phasor by −j rotates the phasor
through −90◦. These facts are used in a.c. theory
since certain quantities in the phasor diagrams
lie at 90◦ to each other. For example, in the R−L
series circuit shown in Fig. 20.8
(a), $\ {V}_{L}$ leads I by
90◦ (i.e. I lags  $\ {V}_{L}$ by 90◦) and may be written as
$\ {jV}_{L}$, the vertical axis being regarded as the imaginary
axis of an Argand diagram. Thus $\ {V}_{R}$ + $\ {jV}_{L}$ =V
and since $\ {V}_{R}$=IR, V =$\ {IX}_{L}$ (where $\ {X}_{L}$ is the inductive
reactance, $\mathrm{2πfL}$ ohms) and V =IZ (where Z is
the impedance) then R+ j XL=Z.
Similarly, for the R−C circuit shown in Fig. 20.8(b),
$\ {V}_{C}$ lags I by 90◦ (i.e. I leads $\ {V}_{C}$ by 90◦) and $ \mathrm {V}_{R} - \mathrm {jV}_{C} = {V}$, from which $ \mathrm {R} - \mathrm {jX}_{C} = {Z}$ (where $\ {X}_{C}$ is the capacitive reactance $\frac{1}{(2 \pi f C)}$ ohms.)
